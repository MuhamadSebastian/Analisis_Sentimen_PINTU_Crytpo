## Cleaning

```python
import re
import string
import nltk
import pandas as pd
nltk.download('wordnet')
nltk.download('punkt_tab')
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

stop_words = set(stopwords.words('indonesian'))
stemmer = StemmerFactory().create_stemmer()
```

```python
# 1. CLEANING
import re

def clean_text(text):
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)  # hapus URL
    text = re.sub(r"[^a-zA-Z\s]", ' ', text)             # hapus karakter non huruf
    text = re.sub(r'\s+', ' ', text).strip()             # hapus spasi berlebih
    return text

df_data['cleaned'] = df_data['content'].apply(clean_text)
```

```python
df_data[['content', 'cleaned']].head(5)
```

```python
# Cek missing value
for col in df_data.select_dtypes(include='object').columns:
    kosong = df_data[col].apply(lambda x: isinstance(x, str) and x.strip() == '').sum()
    print(f"Kolom '{col}' memiliki {kosong} baris kosong (hanya spasi/kosong string)")
```

```python
# Menghapus baris yang memiliki mising value pada kolom 'cleaned'
df_data = df_data[
    (df_data['cleaned'].apply(lambda x: isinstance(x, str) and x.strip() != '')) #&
    #(df_data['casefolded'].apply(lambda x: isinstance(x, str) and x.strip() != ''))
]
```

## Case Folding

```python
# 2. CASE FOLDING
df_data['casefolded'] = df_data['cleaned'].str.lower()
```

```python
df_data[['cleaned', 'casefolded']].head(5)
```

## Tokenizing

```python
# 3. TOKENIZING
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

df_data['tokenized'] = df_data['casefolded'].apply(word_tokenize)
```

```python
df_data[['casefolded', 'tokenized']].head(5)
```

## Normalization

```python
# Buka file Excel
df_excel = pd.read_excel("/content/kamuskatabaku.xlsx")

# Simpan sebagai CSV
df_excel.to_csv("/content/kamuskatabaku.csv", index=False, encoding='latin1')
```

```python
# 5. Normalisasi (WordNetLemmatizer untuk kata baku)
normalized_word = pd.read_csv("/content/kamuskatabaku.csv",
encoding='latin1')
normalized_word_dict = {}
for index, row in normalized_word.iterrows():
  if row[0] not in normalized_word_dict:
    normalized_word_dict[row[0]] = row[1]

def normalized_term(document):
  return [normalized_word_dict[term] if term in
normalized_word_dict else term for term in document]
df_data['normalization'] = df_data['tokenized'].apply(normalized_term)
```

```python
df_data[['tokenized', 'normalization']].head(5)
```

## Stopword Removal

```python
# 4. Stopword Removal
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('indonesian'))

df_data['stopwords'] = df_data['normalization'].apply(lambda words: [w for w in words if w not in stop_words])
```

```python
df_data[['normalization', 'stopwords']].head(5)
```

```python
for word in stop_words:
    print(word)
```

```python
#Cek missing value list
df_data['stopwords'].apply(lambda x: len(x) == 0).sum()
```

```python
# Hapus baris dengan list kosong di kolom 'stopwords'
df_data = df_data[df_data['stopwords'].apply(lambda x: len(x) > 0)].reset_index(drop=True)
```

## Stemming

```python
# 6. STEMMING
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
stemmer = StemmerFactory().create_stemmer()

df_data['stemming'] = df_data['stopwords'].apply(lambda words: [stemmer.stem(w) for w in words])
```

```python
df_data[['stopwords', 'stemming']].head(5)
```

```python
#Menyatukan kembali jadi sebuah kalimat
df_data['stemming'] = df_data['stemming'].apply(lambda x: ' '.join(x))
```

```python
df_data[['stopwords', 'stemming']].head(5)
```

## Labelling Data Positif dan Negatif

```python
pip install textblob
```

```python
from textblob import TextBlob
import pandas as pd

# Fungsi untuk mengambil sentimen, polarity, dan subjectivity
def get_sentiment_details(text):
    analysis = TextBlob(text).sentiment
    polarity = analysis.polarity
    if polarity > 0:
        sentiment = 'positive'
    elif polarity < 0:
        sentiment = 'negative'
    else:
        sentiment = None  # Tandai sebagai None jika netral
    return pd.Series([sentiment, polarity])

# Terapkan fungsi ke kolom 'translated_text'
df_data[['sentiment', 'polarity']] = df_data['translated_text'].apply(get_sentiment_details)

# Hapus baris dengan sentimen netral (None)
df_data = df_data[df_data['sentiment'].notnull()]
```

# Modeling

## TF-IDF Feature Transformation

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X = tfidf_vectorizer.fit_transform(df_data['stemming'])
```

```python
tfidf_features = tfidf_vectorizer.get_feature_names_out()

# Ubah hasil TF-IDF menjadi DataFrame
tfidf_df = pd.DataFrame(X.toarray(), columns=tfidf_features)

# Tampilkan hasil pembobotan
print(tfidf_df)
```

```python
for i in range(tfidf_df.shape[0]):  # jumlah dokumen
    print(f"\nDokumen {i+1}:")
    top_words = tfidf_df.iloc[i].sort_values(ascending=False).head(5)
    for word, score in top_words.items():
        print(f"  {word}: {score:.4f}")
```

```python
# Menjumlahkan nilai TF-IDF dari semua dokumen
total_tfidf = tfidf_df.sum().sort_values(ascending=False)

# Tampilkan 10 kata dengan nilai total TF-IDF tertinggi
print("Kata paling sering muncul berdasarkan skor TF-IDF total:")
print(total_tfidf.head(10))
```

```python
# Jumlah dokumen
jumlah_dokumen = X.shape[0]

# Jumlah kata unik (fitur) setelah seleksi TF-IDF
jumlah_kata_unik = len(tfidf_vectorizer.get_feature_names_out())

print(f"Jumlah dokumen       : {jumlah_dokumen}")
print(f"Jumlah kata unik     : {jumlah_kata_unik}")
```
